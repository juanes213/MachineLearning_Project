```python
df = pd.read_csv('C:/Users/pc/Documents/GitHub/MachineLearning_Project/AAPL_2021_11.txt', names=['Timestamp', 'Price', 'Volume', 'Exchange_code', 'Trade_Condition'], sep=',')
df['Price'] = df['Price'].astype(float)

def Von_Neumann(prices, step_size, threshold):
    up_down = []
    for i in range(0, len(prices) - step_size, step_size):
        if prices[i] < prices[i + step_size]:
            up_down.append(1)
        elif prices[i] > prices[i + step_size]:
            up_down.append(-1)
        else:
            up_down.append(0)
            
    if not up_down:
        return 2
    
    avg = sum(up_down) / len(up_down)
    if -threshold <= avg <= threshold:
        return 0
    return 1 if avg > 0 else -1

def plot_results(prices, signals):
    plt.figure(figsize=(10, 6))
    plt.plot(prices, label='Precio')
    
    buy_signals = np.where(signals == 1)[0]
    sell_signals = np.where(signals == -1)[0]
    
    plt.scatter(buy_signals, prices[buy_signals], marker='^', color='green', label='Compra', alpha=1)
    plt.scatter(sell_signals, prices[sell_signals], marker='v', color='red', label='Venta', alpha=1)
    
    plt.title('Estrategia de Trading con Von Neumann')
    plt.xlabel('Tiempo')
    plt.ylabel('Precio')
    plt.legend()
    plt.show()

prices = df['Price'].values
step_size = 1  
threshold = 0.7

signals = np.array([Von_Neumann(prices[i:i+step_size*2], step_size, threshold) for i in range(0, len(prices)-step_size*2, step_size)])

plot_results(prices, signals)
```
